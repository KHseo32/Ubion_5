```python
offline.columns=['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']
offline
```

```python
online.columns=['기준년월', '성별', '연령대', '총이용건수', '총이용금액']
online
```

```python
online['업종']='온라인'
online
```

```python
df=pd.concat([offline,online], axis=0)
df.info()
```

```python
df['기준년월'] = pd.to_datetime(df['기준년월'], format='%Y%m')
df.info()
```

```python
df2=df.copy()
```

```python
# 성별 컬럼의 데이터 1=남자 2=여자
df2['성별'].replace(
    {
        1: '남자',
        2: '여자',
        9: '무관'
        }
)
df2
```

```python
## map() + (인원함수 | lambda 함수)
def change(x):
    if x ==1:
        return'남자'
    elif x == 2:
        return  '여자'

df2['성별'].map(change)
```

```python
# if 구문을 한줄로 표시 : 참인 경우 return 데이터 if(조건식) else 거짓인 경우 return 데이터
df2['성별'].map(
    lambda x : '여자' if (x==2) else ('남자' if (x==1) else '무관')
)
```

```python
# for 문
df2.reset_index(drop=True, inplace=True)
```

```python

for i in range(len(df)):
    gender = df2['성별'][i]
    if gender == 1:
        df2['성별'][i] = '남자'
    elif gender == 2:
        df2['성별'][i] = '여자'
    else:
        df2['성별'][i] = '무관'
    
```

```python
df2
```

```python
mean_df = df2[['기준년월', '업종', '총이용금액']].groupby(['기준년월', '업종']).sum()

mean_df
```

```python
result_unstacked = mean_df.unstack()


result_unstacked.plot()

plt.ylabel('총이용금액')
plt.title('기간 및 업종별 총 이용 금액')

plt.show()
```
