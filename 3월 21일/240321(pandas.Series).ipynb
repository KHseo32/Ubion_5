### 데이터프레임의 필터링
- loc[인덱스의 조건식, 컬럼의 조건식]
    - 인덱스의 값이나 컬럼의 값을 기준으로 필터링(df.loc[1,'A'])
    - 인덱스의 조건식이나 컬럼의 조건식을 기준으로 필터링
- iloc[인덱스의 위치, 컬럼의 위치]
    - 인덱스의 위치, 컬럼의 위치를 기준으로 필터링(df.iloc[2:4, 0:2])

```python
# index 0의 컬럼 총 사망자의 데이터 출력
corona.loc[0, '총 사망자']
```

```python
# 인덱스 1의 컬럼 총 사망자의 데이터 출력
corona.loc[1, '총 사망자']
```

```python
daily_data = corona.loc[1, '총 사망자'] - corona.loc[0, '총 사망자']
```

```python
# 위의 작업을 corona의 index의 개 수 - 1만큼 반복 실행
data_list = [0]
for i in range(len(corona)-1) : 
    daily_data = corona.loc[i+1, '총 사망자'] - corona.loc[i, '총 사망자']
    data_list.append(daily_data)
```

```python
## 결측치를 특정한 값으로 대체
## 채운다(fill) + 결측치(na) -> fillna()
## fillna(data) : 결측치를 data로 치환
## fillna(method = ) : method 매개변수 -> 'ffill' : 전의 데이터로 치환
                                        # 'bfill' : 후의 데이터로 치환
values = [
    [1,2,3],
    [5,np.nan,np.nan],
    [np.nan,8,9]
]
df = pd.DataFrame(values, columns = ['A', 'B', 'C'])
```

```python
df['B'] = df['B'].fillna(method = 'ffill')
```

```python
df['C']=df['C'].fillna(method = 'bfill')
```

```python
df.fillna(0)
```

```python
## corona 데이터에서 일일 사망자 2, 일일 사망자 3의 결측치 0으로 대체
corona[['일일 사망자2','일일 사망자3']] = corona[['일일 사망자2','일일 사망자3']].fillna(0)
```
