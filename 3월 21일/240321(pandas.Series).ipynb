```python
# 라이브러리 로드
import pandas as pd
```

```python
# 1차원 데이터 형인 Series 생성
pd_series = pd.Series(
    [5000,6000,6500,6500]
)
```

```python
values = [
    [1, 2, 6],
    [3, 4, 1],
    [6, 8, 3]
]
df = pd.DataFrame(values)
```

```python
df
```

```python
## 데이터프레임에는 적어도 3개의 변수가 존재
print('df의 values =', df.values)
print('df의 index =', df.index)
print('df의 columns =', df.columns)
```

```python
# 칼럼의 이름을 변경
df.columns = ['A', 'B', 'C']
```

```python
df
```

```python
## 특정 컬럼의 이름이나 특정 인덱스의 이름을 변경하려면
# rename()
# index 매개변수 : 딕셔너리형태 인자값을 사용 {기준이 되는 인덱스 : 변경할 인덱스 값}
# columns 매개변수 : 딕셔너리형태 인자값을 사용 {기준이 되는 컬럼 : 변경할 컬럼 값}
# inplace 매개변수 : bool형 인자값을 사용(False(기본값) : 데이터 변경 X, True : 데이터 변경)
# df.rename(
#     columns = {'B': 'F'},
#     index = {2 : 5},
#     inplace = True
# )
df= df.rename(
    columns = {'B': 'F'},
    index = {2 : 5}
)
```

```python


df
```

```python
df2 = df.copy()
```

```python
df2.rename(
    columns = {
        'A' : 'X',
        'F' : 'Y',
        'C' : 'Z'
    },
    index = {0 : 'A',
             1 : 'B',
             5 : 'C'
    },
    inplace = True
)
```

```python
print(df2)
```
