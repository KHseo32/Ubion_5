```python
# 라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplot을 이용한 그래프의 시각화
# 라인 그래프 (플롯 그래프)
# y축 데이터만 존재하는 경우
# x축의 데이터는 0부터 시작하여 1씩 증가하는 형태

y_list = [10, 5, 8, 16]
# show() : print() 같은 기능
plt.plot(y_list)
plt.show()
```

```python
# x, y축의 데이터가 모두 존재하는 경우

# 라인 그래프의 라인 색상과 종류
x_list = [1, 3, 6, 10]
y_list = [1, 2, 3, 4]
plt.plot(x_list, y_list, 'r--')
plt.show()
```

```python
## 그래프에 한글 출력
## matplotlib내 폰트 변경 코드 실행
# window와 MAC 사용 폰트 다름


import platform

if platform.system() == 'Darwin':
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.plot([1,2,3,4], [10,20,10,30])
plt.plot([1,2,3,4], [20,40,8,1])
# x축, Y축에 라벨링
plt.xlabel('X')
plt.ylabel('Y')
# 그래프의 이름 표시
plt.title('그래프')
# 범례 추가
plt.legend(['a', 'b'])
plt.show()
```

```python
plt.plot([1,2,3,4], [10,20,10,30], label = 'A')
plt.plot([1,2,3,4], [20,40,8,1], label = 'B')
# x축, Y축에 라벨링
plt.xlabel('X')
plt.ylabel('Y')
# 그래프의 이름 표시
plt.title('그래프')
# 범례 추가
plt.legend(['a', 'b'])
plt.show()
```

```python
# 막대 그래프
x = [1,2,3,4]
y=[80,70,90,75]
plt.barh(x,y)
plt.show()
```

```python
# 옵션 추가(색상, 너비)
plt.bar(x,y, width = 0.2, color = ['r','b','g','y'])
plt.show()
```

```python
# 특정 막대의 색상을 변경
graph = plt.bar(x,y)
graph[2].set_color('r')
plt.show()
```
